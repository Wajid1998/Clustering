## Normalization function

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

## Normalized data frame (considering the numerical part of data)

```python
df_norm = norm_func(crime.iloc[:,1:])
df_norm
```

## create dendrogram

```python
dendrogram = sch.dendrogram(sch.linkage(df_norm, method='complete'))
```

```python
# create clusters
hc = AgglomerativeClustering(n_clusters=4, affinity = 'euclidean', linkage = 'complete')
hc
```

```python
y_hc = hc.fit_predict(df_norm)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
df_norm['h_clusterid'] = Clusters
df_norm
```

```python
df_norm.sort_values('h_clusterid')
```

# KMeans

```python
from sklearn.cluster import KMeans
```

```python
# Normalization function 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_crime_df = scaler.fit_transform(crime.iloc[:,1:])
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_crime_df)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Build Cluster algorithm
clusters_new = KMeans(4, random_state=42)
clusters_new.fit(scaled_crime_df)
```

```python
clusters_new.labels_
```

```python
#Assign clusters to the data set
crime['clusterid_new'] = clusters_new.labels_
```

```python
#these are standardized values.
clusters_new.cluster_centers_
```

```python
crime.groupby('clusterid_new').agg(['mean']).reset_index()
```

# DBSCAN

```python
#Import the libraries
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
crime.info()
```

```python
crime.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
array=crime.values
```

```python
array
```

```python
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array)
X
```

```python
dbscan = DBSCAN(eps=0.8, min_samples=6)
dbscan.fit(X)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
cl
```

```python
pd.concat([crime,cl],axis=1)
```
